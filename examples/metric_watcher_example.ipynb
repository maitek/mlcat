```python
from mlcat.logging import MetricWatcher
from IPython.display import clear_output
import matplotlib.pyplot as plt

with MetricWatcher(watch_vars = ["a", "b"]) as mw:
    for a in range(10):
        b = a**2
        mw.update(scope=locals())
        clear_output(wait=True)
        mw.plot_metrics(["b"])
        plt.legend()
        plt.show()
        plt.pause(0.1)
        mw.print_metrics()
```
